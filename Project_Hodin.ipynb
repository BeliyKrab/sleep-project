```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# A
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv", keep_default_na=False)

print('Первые 5 строк')
display(df.head())
print('Последние 5 строк')
display(df.tail())

print(f'Размер датасета: {df.shape[0]} строк, {df.shape[1]} столбцов')
print('Типы данных столбцов:')
display(df.dtypes)
display(df.describe())

miss_values = df.isnull().sum()
print(f'пропущенные значения\n{miss_values}')

print(f"\nКоличество дубликатов: {df.duplicated().sum()}")
if df.duplicated().sum() > 0:
    df = df.drop_duplicates()
else:
    print('дубликатов нет')

# Преобразование типов
df[['Systolic Pressure', 'Diastolic Pressure']] = df['Blood Pressure'].str.split('/', expand = True).astype(int)
df['BMI Category'] = df['BMI Category'].replace('Normal Weight', 'Normal')

# проверка выбросов
numeric_cols = ['Person ID', 'Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level',
                'Stress Level', 'Heart Rate', 'Daily Steps', 'Systolic Pressure', 'Diastolic Pressure']
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numeric_cols])
plt.title('Boxplot для числовых признаков', fontsize=16)
plt.xticks(rotation=45)
plt.ylabel('Значения')
plt.tight_layout()
plt.show()

# Статистики данных
display(df.describe())
categorical_cols = ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']

for col in categorical_cols:  # уникальные значения
    print(f"Столбец: {col}")
    print(f"Уникальные значения: {df[col].unique()}")
    print(f"Количество уникальных: {df[col].nunique()}")
    print("\nЧастота значений:")
    display(pd.Series(df[col].value_counts()))

# Визуализация корреляций
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Корреляционная матрица числовых признаков')
plt.show()
```

```python
# C. Визуализация данных
# 1 Гистограмма распределения продолжительности сна
plt.figure(figsize=(10, 6))  # размер графика в дюймах
sns.histplot(df['Sleep Duration'], bins=15)
plt.title('Распределение продолжительности сна')
plt.xlabel('Часы сна')
plt.ylabel('Количество людей')
plt.show()
```

```python
# 2. Линейный график: среднее качество сна по возрастам
# Группировка по возрасту
sleep_by_age = df.groupby('Age')['Quality of Sleep'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=sleep_by_age, x='Age', y='Quality of Sleep', marker='o')  # marker='o' - точки на линии
plt.title('Среднее качество сна по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Качество сна (баллы)')

plt.show()
```

```python
# 3. Столбчатая диаграмма: Расстройства сна по профессиям
# Считаем количество расстройств для каждой профессии
disorders_by_occ = df.groupby(['Occupation', 'Sleep Disorder']).size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.barplot(data=disorders_by_occ, x='Occupation', y='count', hue='Sleep Disorder')  # hue - разбивка по цветам
plt.title('Расстройства сна по профессиям')
plt.xlabel('Профессия')
plt.ylabel('Количество людей')
plt.xticks(rotation=45)
plt.legend(title='Расстройство')
plt.show()
```

```python
# 4. Boxplot: Качество сна при разных расстройствах
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sleep Disorder', y='Quality of Sleep')
plt.title('Качество сна при разных расстройствах')
plt.xlabel('Тип расстройства')
plt.ylabel('Качество сна (баллы)')
plt.show()
```

```python
# 5. Heatmap корреляций
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')  # annot=True - цифры в ячейках, fmt=".2f" - 2 знака после запятой
plt.title('Корреляция между числовыми признаками')
plt.show()
```

```python
# D Группировки и агрегации
# 1. Среднее качество сна по профессиям (агрегация = mean)
occupation_quality = df.groupby('Occupation')['Quality of Sleep'].mean().sort_values(ascending=False)
print("\nСреднее качество сна по профессиям:")
print(occupation_quality.head(5))
```

```python
# 2. Подсчет количества расстройств сна (агрегация = count)
disorder_counts = df['Sleep Disorder'].value_counts()
print("\nКоличество расстройств сна:")
print(disorder_counts)
```

```python
# 3. Сводная таблица: Физическая активность по полу и ИМТ агрегация = mean
pivot_table = pd.pivot_table(df, values='Physical Activity Level',
                            index='Gender', columns='BMI Category',
                            aggfunc='mean')
print("\nСредняя физическая активность по полу и ИМТ:")
print(pivot_table)
```

```python
# сохранение
df.to_csv('Sleep_health_processed.csv', index=False)
```
